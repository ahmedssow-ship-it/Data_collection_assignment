### scrape the following informations: brand, model, year, reference, kms driven, adress, price

```python
df = pd.DataFrame()
for index in range(1, 201):
  url = f'https://dakar-auto.com/senegal/motos-and-scooters-3?&page={index}'
  res = get(url)
  soup = bs(res.content, 'html.parser')
  containers = soup.find_all('div', class_ = 'listings-cards__list-item mb-md-3 mb-3')

  data = []
  for container in containers:
    try:
      gen_inf = container.find('h2', 'listing-card__header__title mb-md-2 mb-0').a.text.strip().split()
      gen_inf1 = container.find_all('li', 'listing-card__attribute list-inline-item')
      gen_inf2 = container.find('div', class_ = 'col-12 entry-zone-address').text.strip().replace('\n', '')
      #img_link = container.find('img')['src']
      #brand, model, year
      brand = gen_inf[0]
      model = " ".join(gen_inf[1:len(gen_inf) - 1])
      year = gen_inf[-1]

      #reference, kms driven,  price, adress
      reference = gen_inf1[0].text.split()[-1]
      kms_driven = gen_inf1[1].text.strip().replace('km', '')
      price = "".join(container.find('h3', 'listing-card__header__price font-weight-bold text-uppercase mb-0').text.strip().split()).replace('FCFA', '')
      adress = gen_inf2[:]

      #image
      image = 'https://dakar-auto.com' + container.find('img')['src']

      dic = {'brand': brand,
             'model': model,
             'year': year,
             'reference': reference,
             'kms_driven': kms_driven,
             'price': price,
             'adress': adress,
             'image': image
             }
      data.append(dic)

    except:
      pass

  data_scrape = pd.DataFrame(data)
  df = pd.concat([df, data_scrape], axis = 0).reset_index(drop = True)
```
